```python
# data taken from https://github.com/owid/energy-data
# data taken from https://ourworldindata.org/energy
import pandas as pd
```

```python
energy_data = pd.read_csv("owid-energy-data.csv")
```

```python
#q1 how many unique countries are there in the dataset?
len(set(energy_data["country"]))
```

```python
#q2 how many unique countires in the dataset has a name that starts with A?
all_names = energy_data["country"]
len(set(all_names[all_names.str.startswith("A")]))
```

```python
#q3 make a list of all column names in the dataset with no "_"
all_column_names = list(energy_data.columns)
answer = []
for i in all_column_names:
    if("_" not in i):
        answer.append(i)
answer
```

```python
#q4 make a list containing all column names related to wind
all_column_names = list(energy_data.columns)
answer = []
for i in all_column_names:
    if("wind" in i):
        answer.append(i)
answer
```

```python
#q5 create a data frame containing data for all countires in 2016 that have non-empty(not NaN) gdp
energy_data[(energy_data["year"] == 2016) & (energy_data["gdp"])]
```

```python
#q6 make a list of all country names in 2016 that has a non-negative coal_prod_change_pct value?(return a list)
list(energy_data[(energy_data["year"] == 2016) & (energy_data["coal_prod_change_pct"] >= 0)]["country"])
```

```python
#q7 what is the absolute difference between China's coal_prod_change_pct value in 2013 and 2014(approximate to 3 digit)
china_coal_2013 = float(energy_data[(energy_data["year"] == 2013) & (energy_data["country"] == "China")]["coal_prod_change_pct"])
china_coal_2014 = float(energy_data[(energy_data["year"] == 2014) & (energy_data["country"] == "China")]["coal_prod_change_pct"])
abs(china_coal_2014 - china_coal_2013)
```

```python
#q8 how many countries in 2015 have a wind_energy_per_capita that's larger than 100
len(energy_data[(energy_data["year"] == 2015) & (energy_data["wind_energy_per_capita"] > 100)])
```

```python
#q9 make a list of all countires in 2016 that have a higher nuclear_share_energy than oil_share_energy
energy_data[(energy_data["year"] == 2016) & (energy_data["nuclear_share_energy"] > energy_data["oil_share_energy"])]
```

```python
#q10 what is float vlaue of the World total coal_consumption in 2017
float(energy_data[(energy_data["year"] == 2017) & (energy_data["country"] == "World")]["coal_consumption"])
```
